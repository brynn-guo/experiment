```python
import random
import string
import mysql
import mysql.connector
from datetime import datetime, timedelta
# import tokens
from importlib import reload
import pandas as pd
import numpy as np
from quafu import QuantumCircuit as quafuQC

import sys

sys.path.append('/Users/brynn/work/gyb')

import gates
import tools
```

```python
from quark import Task
token = 'p2wHCXIXyQULIwHT:WEPXBeXyDZJuYNxVbTG2ji26iJ/1oO1lUN4dkM{NENzFkO4J{OypkJxiY[jxjJ{VkP{FkPxJEJxNUM5BUM1JENzJjPjRYZqKDMj53ZvNXZvNYbyGnZBKXfwW4[jpkJzW3d2Kzf'
tmgr = Task(token)
```

```python
def series(type,qs):

    qlisp = []
    circ = quafuQC(2)
    
    if type == 0:
        gates.add_measure(qlisp, circ,[0,1])
    if type == 1:
        gates.add_R('Rx', qlisp, circ, np.pi, 0)
        gates.add_measure(qlisp, circ,[0,1])
    if type == 2:
        gates.add_R('Rx', qlisp, circ, np.pi, 1)
        gates.add_measure(qlisp, circ,[0,1])

    if type == 3:
        gates.add_R('Rx', qlisp, circ, np.pi, 0)
        gates.add_R('Rx', qlisp, circ, np.pi, 1)
        gates.add_measure(qlisp, circ,[0,1])


    if type == 4:
        gates.add_cnot(qlisp, circ,  0, 1)
        gates.add_measure(qlisp, circ,[0,1])

    if type == 5:
        gates.add_H(qlisp, circ,  0)
        gates.add_H(qlisp, circ,  1)
        gates.add_cnot(qlisp, circ,  0, 1)
        gates.add_H(qlisp, circ,  0)
        gates.add_H(qlisp, circ,  1)
        gates.add_measure(qlisp, circ,[0,1])
        
    qmap = qmap = {0:qs[0], 1:qs[1]}
    qlisp = tools.map_qlisp(qlisp,qmap)

    return qlisp, circ
```

```python
qs_list = [('Q2', 'Q3'), # 99.78, 99.93, 98.0
           ('Q11', 'Q12'), # 99.819, 99.938, 99.9
           ('Q16','Q17'), # 99.946, 99.941, 99.9
           ('Q18', 'Q19'), # 99.88, 99.921, 99.0
           ('Q58', 'Q59'), # 99.924, 99.937, 99.19
           ('Q59', 'Q60'), # 99.937, 99.934, 98.95
           ('Q81', 'Q82'), # 99.931, 99.94, 99.14
           ]
```

```python
series(type = 5 , qs = qs_list[2])[1].plot_circuit()
```

```python
from quafu import simulate
qc = series(type = 4 , qs = qs_list[2])[1]
simulate(qc).plot_probabilities()
```

```python
tid_df = pd.DataFrame(index = qs_list, columns= range(6))

for qs in tid_df.index:
    for type in tid_df.columns:
        qlisp, circ = series(type = type, qs = qs)
        task = {
                'chip':'Baihua',
                'name':'gyb_read',
                'circuit':qlisp,
                'compile' : False, 
            }
        tid = tmgr.run(task, repeat=10)
        tid_df[type][qs] = tid
```
